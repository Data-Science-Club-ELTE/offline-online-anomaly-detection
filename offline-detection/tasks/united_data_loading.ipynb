

Combines the data loading approaches from **Alessandro** and **Nazrin**:
- **Alessandro**: Kaggle download via `kagglehub`, load CSV, report **shape** (rows, columns).
- **Nazrin**: Same source and load, plus **column names**, **first 5 rows**, and **dataset info** (dtypes, non-null counts).

Single place to load the dataset for the rest of the offline_team pipeline.

```python
import kagglehub
import pandas as pd
from pathlib import Path


dataset_path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")
csv_file = Path(dataset_path) / "creditcard.csv"


df = pd.read_csv(csv_file)


print("Shape of the dataset:", df.shape)
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

print("\nColumn Names:")
print(df.columns.tolist())

print("\nFirst 5 Rows:")
display(df.head())

print("\nDataset Info (dtypes, non-null):")
df.info()
```
