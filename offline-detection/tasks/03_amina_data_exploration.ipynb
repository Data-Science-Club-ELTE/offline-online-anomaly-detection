```python
import kagglehub
import pandas as pd
from pathlib import Path

dataset_path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")
csv_file = Path(dataset_path) / "creditcard.csv"
df = pd.read_csv(csv_file)

print("Loaded shape:", df.shape)
```



Inspect dtypes, non-null counts, and memory usage.

```python
print("Data types and nullability:")
df.info()
```

```python

dtype_summary = pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.astype(str),
    "non_null": df.notna().sum(),
    "null_count": df.isna().sum()
})
display(dtype_summary)
```



Numeric summary: count, mean, std, min, quartiles, max.

```python
df.describe()
```

```python

df.describe(include="all", percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])
```

```python

print("Class distribution:")
print(df["Class"].value_counts(dropna=False))
print("\nProportions:")
print(df["Class"].value_counts(normalize=True))
```
