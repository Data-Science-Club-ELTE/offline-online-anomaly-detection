```python

import kagglehub
import pandas as pd
from pathlib import Path


dataset_path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")
csv_file = Path(dataset_path) / "creditcard.csv"


df = pd.read_csv(csv_file)

print("Shape of the dataset:", df.shape)
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

print("\nColumn Names:")
print(df.columns.tolist())

print("\nFirst 5 Rows:")
print(df.head())

print("\nDataset Info (dtypes, non-null):")
df.info()
```

```python
df
```

## Class Distribution

```python
class_counts = df['Class'].value_counts()
```

```python
class_percentage = df['Class'].value_counts(normalize=True) * 100
```

```python
print(f"Non-Fraud (0): {class_counts[0]} ({class_percentage[0]:.2f}%)")
print(f"Fraud (1): {class_counts[1]} ({class_percentage[1]:.2f}%)")
```

## Distribution Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.distplot(df[df['Class'] == 0]['Amount'], label='Normal', hist=False)
sns.distplot(df[df['Class'] == 1]['Amount'], label='Fraud', hist=False)
plt.title('Distribution of Transaction Amount by Class')
plt.xlim([-100, 3000])
plt.legend()
plt.show()
```

1) Both classes are heavily directed toward low amounts.
2) High overlap suggests "Amount" alone is not a primary discriminator for fraud.
